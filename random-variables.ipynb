## Probability Mass Function

probability mass function (pmf) is a function that maps from each possible value of a random variable to its probability.

for example, the pmf of a coin flip is: 

$$ P(X = x) = \begin{cases} 0.5 & \text{if } x = 0 \text{ or } x = 1 \\ 0 & \text{otherwise} \end{cases} $$


## Expectation

A random variable is fully represented by its probability mass function. from the pmf, we can compute the expectation of a random variable, which is the mean of the distribution. 

$$ E[X] = \sum_{x} x P(X = x) $$

expectation is just a weighted average of the possible values of the random variable, where the weights are the probabilities.



## Variance

variance is a measure of how spread out the values of a random variable are.

$$ Var[X] = E[(X - E[X])^2] $$

while expection gives you the weighted average of the values, variance shows you how spread out the values are.

"the rich gettin' richer and the poor gettin' poorer." this means variance is getting bigger and bigger. 


## Standard Deviation

variance is good at telling us how spread out the values are, but it is hard to interpret because the units are the square of the original units. 

tha's why we use standard deviation, which is the square root of variance.


## Distribution

there are a few commnly used distribution models:

they are all defined by a set of parameters. think of them as a family of distributions, with pre-derived pmf, expectation, variance and other properties, that you can use to simulate different kinds of data, all you need is to provide parameters.

it's like a template in C++

* Bernoulli distribution. it is a special case of the binomial distribution, where n = 1. it is a distribution over a single coin flip. the parameter is p, the probability of heads.

* Binomial distribution. it is the distribution over the number of heads in n coin flips. the parameters are n and p, the number of coin flips and the probability of heads.

* Poisson distribution. it is the distribution over the number of events in a fixed interval of time. the parameter is Î», the average number of events per interval.
